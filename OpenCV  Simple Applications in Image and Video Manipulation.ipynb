5 - How to Read, Write, Show Videos from Camera in OpenCV عربي

```python
import cv2

# فتح الكاميرا
cap = cv2.VideoCapture(0)

# إعداد فيديو الكاتب
fourcc = cv2.VideoWriter_fourcc(*"XVID")
out = cv2.VideoWriter("out.avi", fourcc, 20.0, (640, 480))

print(cap.isOpened())

# التحقق من أن الكاميرا تعمل
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

        # كتابة الإطار في ملف الفيديو
        out.write(frame)

        # تحويل الإطار إلى صيغة رمادية
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow("frame", gray)

        # الخروج عند ضغط "q"
        if cv2.waitKey(1) == ord("q"):
            break
    else:
        break

# تحرير الموارد
cap.release()
out.release()
cv2.destroyAllWindows()
```

9 - Show Date and Time on Videos using OpenCV Python عربي


```python
import cv2
import datetime
cap = cv2.VideoCapture(0)
#print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
#print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
cap.set(3, 3000)
cap.set(4, 3000)
print(cap.get(3))
print(cap.get(4))
while(cap.isOpened()):#open camera
    ret, frame = cap.read() #read frames         
    if ret == True:#if there is a frame         

       font = cv2.FONT_HERSHEY_SIMPLEX             
       text = 'Width: '+ str(cap.get(3)) + ' Height:' + str(cap.get(4))             
       datet = str(datetime.datetime.now())             
       frame = cv2.putText(frame, text, (10, 50), font, 1,             
                           (0, 255, 255), 2)              
       frame = cv2.putText(frame, datet, (10, 100), font, 1,             
                           (0, 255, 255), 2)             
       cv2.imshow('frame', frame)             

       if cv2.waitKey(1)== ord('q'):             #wait key from keyboard             
         break                   
    else:      #if ret =false         
        break               

cap.release()# close camera
cv2.destroyAllWindows() #close window
```

10 - Handle Mouse Events in OpenCV عربي

```python
import numpy as np
import cv2

events = [i for i in dir(cv2) if "EVENT" in i]
print(events)

def click_event(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print(x, "", y)
        font = cv2.FONT_HERSHEY_SIMPLEX
        strXY = str(x) + "," + str(y)
        cv2.putText(img, strXY, (x, y), font, 0.5, (255, 225, 0), 2)
        cv2.imshow("image", img)
    elif event == cv2.EVENT_RBUTTONDOWN:
        blue = img[y, x, 0]
        green = img[y, x, 1]
        red = img[y, x, 2]
        font = cv2.FONT_HERSHEY_SIMPLEX
        strBGR = f"{blue}, {green}, {red}"
        cv2.putText(img, strBGR, (x, y), font, 0.5, (0, 255, 255), 2)
        cv2.imshow("image", img)

# Load an image
img = cv2.imread('F:\\16.jpg')

cv2.imshow('image', img)
cv2.setMouseCallback('image', click_event)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

How to Bind Trackbar To OpenCV Windows عربي


```python
import cv2 as cv
import numpy as np

# دالة لتُستخدم كـ Callback عند تغيير قيمة التراك بار
def nothing(x):
    pass

# إنشاء صورة سوداء بمقاس 300x512 بكسل وثلاثة قنوات للألوان
img = np.zeros((300, 512, 3), np.uint8)
cv.namedWindow('image')

# إنشاء التراك بار للألوان (الأزرق، الأخضر، الأحمر)
cv.createTrackbar('B', 'image', 0, 255, nothing)
cv.createTrackbar('G', 'image', 0, 255, nothing)
cv.createTrackbar('R', 'image', 0, 255, nothing)

# إنشاء تراك بار للتحكم في تشغيل/إيقاف الألوان
switch = '0 : OFF\n1 : ON'
cv.createTrackbar(switch, 'image', 0, 1, nothing)

# حلقة العرض والتحديث
while True:
    cv.imshow('image', img)
    
    # الخروج من الحلقة إذا تم الضغط على مفتاح "Esc"
    k = cv.waitKey(1) & 0xFF
    if k == 27:  # مفتاح Esc
        break

    # قراءة قيم التراك بار للألوان
    b = cv.getTrackbarPos('B', 'image')
    g = cv.getTrackbarPos('G', 'image')
    r = cv.getTrackbarPos('R', 'image')
    s = cv.getTrackbarPos(switch, 'image')

    # ضبط الألوان حسب حالة المفتاح
    if s == 0:  # إذا كانت OFF، خلي الخلفية سوداء
        img[:] = 0
    else:       # إذا كانت ON، خلي الخلفية باللون المحدد
        img[:] = [b, g, r]

# إغلاق جميع النوافذ عند الخروج من الحلقة
cv.destroyAllWindows()
```
